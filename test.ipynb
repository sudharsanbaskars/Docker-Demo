```python
from sklearn.preprocessing import StandardScaler
from sklearn.impute import KNNImputer
import numpy as np
import pandas as pd
from sklearn import decomposition
import matplotlib.pyplot as plt
```

```python
train_path = "Prediction_FileFromDB/InputFile.csv"
```

```python
train_df = pd.read_csv(train_path)
```

```python
train_df.shape
```

```python
train_df.head()
```

```python
def impute(data):
    imputer=KNNImputer(n_neighbors=3, weights='uniform',missing_values=np.nan)
    new_array=imputer.fit_transform(data) # impute the missing values
    # convert the nd-array returned in the step above to a Dataframe
    new_data=pd.DataFrame(data=new_array, columns=data.columns)
    return new_data
```

```python
train_df = impute(train_df)
```

```python
train_df = train_df.drop("Wafer", axis="columns")
```

```python
#X = train_df.drop("Output", axis="columns")
#y = train_df["Output"]
```

```python
standardized_data = StandardScaler().fit_transform(train_df)
print(standardized_data.shape)
```

```python
pca = decomposition.PCA(n_components=163)
pca_data = pca.fit_transform(standardized_data)
```

```python
pca_data.shape
```

```python
pca_df = pd.DataFrame(pca_data)
```
